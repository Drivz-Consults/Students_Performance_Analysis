```python
#To use imputer method for missing values, it is right to split the data into train and test.
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
#Then call out the columns that are floats of intega with missing values.
cols = ["lunch","test_preparation_course","math_score","reading_score","writing_score","science_score","total_score"]
```

```python
studp = pd.read_csv("Student_performance_10k.csv",usecols = cols)
studp
```

```python
studp.info()
```

```python
studp['math_score'] = pd.to_numeric(studp['math_score'], errors='coerce')
```

```python
studp.info()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    studp.drop('total_score', axis=1), 
    studp['total_score'], test_size=0.3, random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
#Checking for the percentage of missing values in the trainsplit set:
x_train.isnull().mean()
```

```python
#Using the median imputation transformer from sikiklearn to fix the missing values in the trainsplit set:
imputer = SimpleImputer(strategy='median')
```

```python
#Let's fix the median strategy and see what the median will be for each column.
imputer.fit(x_train)
imputer.statistics_
```

```python
#Let's infer the knowledge of the xtrain on the xtest as well to fix the missing values in the columns
x_train_fixed = imputer.transform(x_train)
x_test_fixed = imputer.transform(x_test)
```

```python
x_train_df = pd.DataFrame(x_train_fixed)
x_train_df
```

```python
x_test_df = pd.DataFrame(x_test_fixed)
x_test_df
```
